First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.

Questions and Findings:

Are all launch sites in proximity to the Equator line?

Finding: The map shows that SpaceX launch sites are not uniformly located near the Equator. For example:
Cape Canaveral Space Force Station (CCAFS SLC 40) and Kennedy Space Center (KSC LC 39A) are located in Florida, USA, relatively far from the Equator (latitude around 28.5°N).
Vandenberg Space Force Base (VAFB SLC 4E) is situated in California, USA, also far from the Equator (latitude around 34.7°N).
Guiana Space Centre (CSG) is much closer to the Equator (latitude around 5.2°N), reflecting its strategic advantage for launching geostationary satellites.
Explanation: Launch sites are strategically positioned considering various factors, including the type of missions they support. Sites closer to the Equator, like the Guiana Space Centre, benefit from the Earth's rotational speed to gain additional velocity for launches into geostationary orbits. Other sites, like those in the US, are often chosen for their security, infrastructure, and logistical considerations rather than proximity to the Equator.

Are all launch sites in very close proximity to the coast?

Finding: Most of the SpaceX launch sites are located near the coast:
CCAFS SLC 40 and KSC LC 39A are near the coast of Florida.
VAFB SLC 4E is located on the California coast.
Guiana Space Centre (CSG) is situated on the coast of French Guiana.
Explanation: Launch sites are often placed near coastlines to ensure safety. Launching rockets over the ocean minimizes the risk to populated areas in case of a launch failure or rocket malfunction. Coastal locations also facilitate the use of open sea as a safe area for rockets that might not reach their intended orbit or have to be destroyed for safety reasons.

Summary
The visualization confirms that while not all SpaceX launch sites are close to the Equator, most are positioned near the coast to enhance safety and leverage geographical advantages for different types of missions. The strategic placement of these sites reflects a balance between safety, logistical needs, and mission requirements.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Approximate radius of earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Coordinates for the launch site
launch_site_coords = [28.563197, -80.576820]  # Example: CCAFS SLC-40

# Coordinates for the closest point of interest
point_of_interest_coords = [28.564000, -80.578000]  # Example: Closest city or railway

# Calculate distance
distance = calculate_distance(
    launch_site_coords[0], launch_site_coords[1],
    point_of_interest_coords[0], point_of_interest_coords[1]
)

# Create a folium map centered around the launch site
site_map = folium.Map(location=launch_site_coords, zoom_start=15)

# Add the launch site marker
folium.Marker(
    location=launch_site_coords,
    popup='Launch Site',
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Add marker for the point of interest with distance information
folium.Marker(
    location=point_of_interest_coords,
    icon=folium.DivIcon(
        icon_size=(200, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
    )
).add_to(site_map)

# Draw a PolyLine between the launch site and the point of interest
folium.PolyLine(
    locations=[launch_site_coords, point_of_interest_coords],
    weight=2,
    color='blue',
    opacity=0.8
).add_to(site_map)

# Add MousePosition to get coordinates while exploring
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Display the map
site_map
```
